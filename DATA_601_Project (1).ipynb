# **Project Final Report: Mercury in Fish**

Mercury is a naturally occurring heavy metal found in Earth's crust, that can exist in three forms: organic, inorganic, and elemental (Bernhoft, 2012). Human exposure primarily occurs through methyl-mercury, an organic form where mercury is bonded to carbon atoms (Bernhoft, 2012). This is also the form of mercury that is known to have the highest toxicity to humans (Hong et al., 2012). Methylmercury is formed when mercury enters aquatic systems like lakes and rivers (Ulrich et al., 2001). First, both natural sources (e.g., volcanic emissions, rock weathering) and human activities (e.g., coal combustion, mining) release mercury into the atmosphere (Bernhoft, 2012). Mercury can then travel long distances before converting into a soluble form and settling back to Earth through precipitation (Clarkson et al., 2003). Once in aquatic environments, microorganisms convert inorganic mercury into methyl-mercury by adding a CH3 (methyl) group to it (Bernhoft, 2012). The methylmercury can then be absorbed by fish either through water or through their diet, as they consume smaller organisms that have already accumulated this toxic compound. This bioaccumulation process leads to higher concentrations of methyl-mercury in larger, predatory fish, which are often consumed by humans (Clarkson et al., 2003). Methylmercury has long been known to be a potent neurotoxin and reproductive toxin (Clarkson et al., 2003). Upon consumption of contaminated fish methyl-mercury is rapidly and efficiently absorbed via the intestine, and from there it is able to enter the bloodstream (Bernhoft, 2012). In adults, the primary impacted tissue is the brain. The presentation of methyl-mercury toxicity in adults consists of issues with balance and vision (Clarkson et al., 2003). However, a much more intense presentation is seen through prenatal exposure, as the fetal brain is more susceptible (Clarkson et al., 2003). Once it enters the mothers body methyl-mercury can cross the placental barrier thereby impacting the fetus (Clarkson et al., 2003). Furthermore, studies have found that mothers that present as asymptomatic or with mild toxicity can have children that display severe neurological dysfunction (Hong et al., 2012). Health impacts due to high dose prenatal exposure include but are not limited to seizures, mental retardation, growth disorders, limb deformities and cerebral palsy (Hong et al., 2012).

While this issue may seem irrelevant to individuals residing in a developed country, a study conducted by the Canadian Health Measures Survey from 2007 to 2009 found that 2.2% of women of childbearing age had blood mercury levels above the recommended threshold (Lye et al., 2013). While 2.2% may seem like a negligible amount, this number is significant when considering that exposure to methyl-mercury can cause irreversible harm to developing fetuses. Additionally, due to both cultural practices and food insecurity many indigenous communities in Alberta rely on fish as a staple part of their diet, which puts them at an elevated risk of methyl-mercury exposure (Marushka et al., 2021). These factors underscore the need for a comprehensive understanding of methyl-mercury exposure and factors that contribute to it.

## **Data Cleaning & Data Wrangling**

To address each guiding question comprehensively, thorough data cleaning was essential. First, a copy of the original dataframe was created to preserve the unmodified dataset for potential future use. Only relevant columns were retained for analysis, and unnecessary ones were dropped to streamline the dataset. Specifically, the following columns were removed: Program ID, Sample ID, Date Flag, Weight Measurement, Maturity, Aging Structure, FWMIS ID, Species Code, Length Estimate Factor, Fork Length Measurement, Total Length Measurement, and Age (years).

```python
# removing the columns we do not need
df.drop(['Program_ID', 'Sample_ID','Date_flag','WeightMeasurement', 'Maturity', 'Aging Structure',
         'FWMIS ID', 'Species Code','LengthEstimateFactor', 'Fork LengthMeasurement ', 'Total LengthMeasurement', 'Age (years)' ], axis = 1, inplace = True)
```

```python
# check list of columns after removing the columns we do not need
df.columns
```

Next, data types were converted appropriately. The remaining columns included: Collection Date, Waterbody Type, Waterbody Name, Land Use Region, Common Name, Binomial Name, Longitude, Latitude, Sex, Fork Length (mm), Weight (g), Total Length (mm), and Hg (mg/kg). The Collection Date was converted into a datetime object, while categorical variables like Waterbody Type, Waterbody Name, Land Use Region, Common Name, Binomial Name, and Sex were transformed into categorical data types. Numerical columns such as Latitude, Longitude, Fork Length (mm), Weight (g), Total Length (mm), and Hg (mg/kg) were converted to floats.


```python
# strip any whitespace from column titles for ease
df.columns = df.columns.str.strip()

# some latitude and long enteries have two enteries in parantheses, lets get rid of what is in parantheses
df['Latitude'] = df['Latitude'].str.replace(r'\s*\(.*\)', '', regex=True).str.strip()
df['Longitude'] = df['Longitude'].str.replace(r'\s*\(.*\)', '', regex=True).str.strip()
df['Collection Date'] = pd.to_datetime(df['Collection Date'], dayfirst= True, format = 'mixed')
```

```python
#convert column to numeric types
numerical_columns = ['Fork Length (mm)', 'Hg (mg/kg)', 'Latitude', 'Longitude', 'Total Length (mm)', 'Weight (g)' ]
df[numerical_columns] = df[numerical_columns].apply(pd.to_numeric, errors='coerce')

#clean and covert category columns
categorical_columns = ['Waterbody Type', 'Land Use Region', 'Common Name', 'Binomial Name', 'Sex', 'Waterbody Name']
df[categorical_columns] = df[categorical_columns].apply(lambda col: col.str.strip().astype('category'))

#sort the dataframe by collecting date
df.sort_values(by = ['Collection Date'], inplace = True)
df.dropna(subset=['Hg (mg/kg)'], inplace = True)
```

A column for feeding behavior was added, classifying each fish as a predator, prey, or both. This data was sourced from the Alberta government website, and lists were created for each feeding category. These lists were then added to a dictionary, using species names as keys and feeding behaviors as values. The map method was employed to integrate this column into the dataframe. Furthermore, the dataset contained several NaN values, and a strategy was developed to handle them. To maintain data integrity, rows with missing values were not omitted, except for those with NaN values in the Mercury (Hg) column, which were dropped due to their relevance to the analysis.


```python
# add a column for whether each species is a predator or prey
predators = [
    "Brook Trout",
    "Brown Trout",
    "Burbot",
    "Cutthroat Trout",
    "Goldeye",
    "Lake Trout",
    "Mooneye",
    "Northern Pike",
    "Rainbow Trout",
    "Sauger",
    "Walleye",
    "Yellow Perch"
]
prey = [
    "Lake Chub (Cisco)",
    "Cisco",
    "Flathead Chub",
    "Lake Chub",
    "Lake Whitefish",
    "Longnose Dace",
    "Longnose Sucker",
    "Shorthead Redhorse",
    "Trout Perch",
    "White Sucker"
]

both  = [
    "Mountain Whitefish",
    "Yellow Perch"
]
feeding = {}
for fish in predators:
    feeding[fish] = "Predator"
for fish in prey:
    feeding[fish] = "Prey"
for fish in both:
    feeding[fish] = "Predator and Prey"
color_map = {
    'Predator': 'skyblue',
    'Prey': 'lightcoral',
    "Predator and Prey": "lightgreen"
}

df["Feeding Behavior"] = df["Common Name"].map(feeding)
```

Several dataframes were created to explore statistics across different categories. This approach was utilized to provide a clear way to answer the guiding questions and facilitate the creation of visualizations. A summary of mercury concentrations was generated using the describe method, with an additional column added for the range, as it was part of the guiding questions. Another dataframe was created to explore how various metrics varied across each fish species, named "averages". Specifically it included average weight, fork length, total length, mercury concentration, and feeding behavior for each species.


```python
# first averages of each variable of interest by species
averages = df.groupby('Common Name').agg({
    'Hg (mg/kg)': 'mean',
    'Weight (g)': 'mean',
    'Total Length (mm)': 'mean',
    'Fork Length (mm)': 'mean',
    "Feeding Behavior":"first"}).reset_index()
averages.columns = ['Species', 'Average Hg (mg/kg)', 'Average Weight (g)', 'Average Total Length (mm)', 'Average Fork Length (mm)', "Feeding Behavior"]
averages.sort_values(by='Average Hg (mg/kg)', ascending=False, inplace = True)
```

```python
# some general mercury stats
mercury_general_stats = (df['Hg (mg/kg)'].describe())
mercury_general_stats['Range'] = df['Hg (mg/kg)'].max() - df['Hg (mg/kg)'].min()
mercury_general_stats.rename(index={'50%': 'Median'}, inplace=True)
mercury_mode = df['Hg (mg/kg)'].mode()
mercury_general_stats['Mode'] = mercury_mode[0]
```

Next, a custom function was defined to generate additional dataframes, grouped by various columns (e.g., waterbody type, waterbody name, species name, sex, land use region) to calculate the average, median, mode, range, and standard deviation of mercury levels across each of the explanatory variables. Additionally, the data frame containing the information for waterbody name was modified to include latitude and longitude for each waterbody, to create a map seen later in the analysis.


```python
def calculate_mercury_stats(df, group_by_column):
    ''' Function that calculates mercury statistics by grouping for each column of interest'''
    mercury_stats = df.groupby(group_by_column, observed=False).agg({
        'Hg (mg/kg)': [
            'mean',
            'std',
            'median',
            lambda x: x.mode().iloc[0],
            lambda x: x.max() - x.min()]})
    mercury_stats.columns = ['Mean Hg (mg/kg)', 'Std Dev Hg (mg/kg)', 'Median Hg (mg/kg)', 'Mode Hg (mg/kg)', 'Range Hg (mg/kg)']
    return mercury_stats.reset_index()

# grouping of how mercury varies by each species
mercury_stats_by_species = calculate_mercury_stats(df, 'Common Name')

# grouping of how mercury varies by each waterbody name
mercury_stats_by_waterbody = calculate_mercury_stats(df, 'Waterbody Name')

# grouping of mercury stats by sex
mercury_stats_by_sex = calculate_mercury_stats(df, 'Sex')

# grouping of mercury stats by the waterbody type
mercury_stats_by_WB_type = calculate_mercury_stats(df, 'Waterbody Type')

# grouping of mercury stats by land_use_region
mercury_stats_by_land_use_region = calculate_mercury_stats(df, 'Land Use Region')
```

```python
# sort mercury_stats_by_waterbody dataframe by the mean values in descending order
mercury_stats_by_waterbody.sort_values(by='Mean Hg (mg/kg)', ascending=False, inplace = True)

# create mercury_stats_by_waterbody_with_coord dataframe
mercury_stats_by_waterbody_with_coord = df.groupby('Waterbody Name', observed=False).agg({
    'Hg (mg/kg)': ['mean'],
    'Latitude': 'first',
    'Longitude': 'first'
}).reset_index()
mercury_stats_by_waterbody_with_coord.columns = ['Waterbody Name', 'Mean Hg (mg/kg)', 'Latitude', 'Longitude']

# sort mercury_stats_by_waterbody_with_coord dataframe by the mean values in descending order
mercury_stats_by_waterbody_with_coord.sort_values(by='Mean Hg (mg/kg)', ascending=False, inplace = True)
```

Lastly, dataframes were developed to analyze how mercury concentrations varied by season and year. A dictionary mapping months to seasons was created, and a new column was added to the dataframe, reflecting seasonal information. This was done by extracting the month from the collection date and mapping it to the appropriate season. The custom function was then applied to calculate mercury statistics for each season. Similarly, the year was extracted from the collection date, and mercury statistics were computed for each year.


```python
# create month_to_season_dict dictionary to map months to seasons
month_to_season_dict = {
    12: 'Winter', 1: 'Winter', 2: 'Winter',
    3: 'Spring', 4: 'Spring', 5: 'Spring',
    6: 'Summer', 7: 'Summer', 8: 'Summer',
    9: 'Fall', 10: 'Fall', 11: 'Fall'
}

# add 'Season' column with the extracted months and map them to seasons
df['Season'] = df['Collection Date'].dt.month.map(month_to_season_dict)

# calculate mercury stats by season
mercury_stats_by_season = calculate_mercury_stats(df, 'Season')

# add 'Year' column with the extracted year
df['Year'] = df['Collection Date'].dt.year

# calculate mercury stats by year
mercury_stats_by_year = calculate_mercury_stats(df, 'Year')
```

#### ***What are the Correlations between the Size (Length and Weight) and Mercury Concentration?***

To explore how fish size relates to mercury concentrations, we first filtered the original dataset to include only the 'Total Length', 'Weight', and 'Hg (mg/kg)' columns. Any unknown or infinite values were removed, and the remaining data was converted into numeric format for accurate calculations, this process was repeated for each of the correlation calculations detailed below for qualitative variables.

A pair plot was then generated, allowing us to visualize the relationships between length, weight, and mercury concentration. The plot shows clustered data points for both length vs. mercury concentration and weight vs. mercury concentration, suggesting a possible influence of fish size on mercury levels.

To confirm this relationship we calculated the Pearson correlation coefficient, using the SciPy stats module from the SciPy library. This coefficient measures the strength and direction of a linear relationship between two continuous variables. Values close to 1 or -1 indicate a strong relationship. The Pearson correlation coefficients for both relationships were found to be approximately 0.3, indicating moderate positive correlations. Additionally, the p-values were found to be significant which means there is a statistically significant relationship between fish size (length and weight) and mercury concentrations. A p-value less than 0.05 suggests that the likelihood of observing such a correlation by chance is very low, lending credibility to our findings. This means we can be reasonably confident that as the size of the fish increases, there is a tendency for mercury concentrations to rise as well.

Based on these results, it can be concluded that fish size, in terms of both length and weight, is associated with increased mercury content. As a  supplement, Ordinary Least Squares was also utilized to further explore the correlation between length and weight. The OLS analysis revealed that total length had a coefficient of 0.0006, meaning that for every millimeter increase in length, mercury concentration increases by 0.0006 mg/kg. Weight had a coefficient of 1.76 x 10^-5, indicating that for every gram increase in weight, mercury concentration increases by 0.0000176 mg/kg. Both predictors were statistically significant with p-values of 0.000 for length and 0.013 for weight. Overall, these findings confirm that fish size is associated with increased mercury content. Although the model’s explanatory power is low, it still demonstrates that both length and weight have significant positive impacts on mercury concentration levels.

#### ***What is the Correlation between Sex of the Fish and Mercury Concentration?***

# **Discussion and Conclusion**

## **Process and Results**

Our data led to several key conclusions. Firstly, we observed an increase in mercury levels in fish over time, which aligned with our initial expectations. Methylmercury forms when atmospheric mercury, from both anthropogenic sources (e.g., coal combustion, mining) and natural sources (e.g., volcanic emissions, rock weathering), is deposited into water bodies (Bernhoft, 2012). Given the rise in industrial activities, we anticipated this upward trend in mercury levels. However, the Government of Canada reports an 85% reduction in mercury emissions to air and a 50% reduction in water since 1990 (Government of Canada, 2023). This suggests that while our findings matched expectations, other factors may be at play. The Government of Canada notes that elevated mercury levels could be due to the re-volatilization of previously deposited mercury, rising water temperatures from climate change, which accelerate methylmercury production, and the acidification of aquatic ecosystems (Government of Canada, 2023).

Next, we identified several fish characteristics that correlate with mercury levels. Predatory species had higher average mercury concentrations than prey, and larger fish, in terms of both weight and length, exhibited higher mercury levels as well. These relationships were supported by correlations showing that fish size, length, and weight were strong predictors of higher mercury concentrations. The relationship between predator-prey and mercury levels was expected  due to bioaccumulation,  where predators accumulate mercury by consuming contaminated prey (Government of Canada, 2023). Similarly, we anticipated finding higher mercury levels in larger fish since predators are typically large species, further supporting bioaccumulation. Additionally the correlation observed between fish size, namely length and weight, and mercury levels were also expected as these are well documented in the literature (Holbert et al., 2023). Particularly, it has been shown that larger fish tend to accumulate more mercury as they are usually predators, and live longer, resulting in extended periods of bioaccumulation (Holbert et al., 2023).

The species was also found to be a factor that was correlated with mercury concentration. Average mercury concentration varied widely between different species. Particularly Sauger, Walleye, and Northern Pike, exhibited notably higher mercury levels. Once again, this was expected, as these species are considered high-risk for mercury contamination (Government of Alberta, 2009). In contrast, species such as Lake Whitefish and Yellow Perch consistently fall within the low-risk category, with mercury levels remaining below the safe threshold.  Additionally, we found that female fish had higher average mercury concentrations than males, and statistical tests confirmed a significant difference between the two. This result was unexpected, as we explored sex differences without anticipating a relationship. However, this finding may be explained by the fact that female fish had higher average weights, which we found to be correlated to mercury concentration.

We found that mercury levels peaked in spring, followed by fall, summer, and winter. Initially, we hypothesized that summer would exhibit the highest concentrations as a result of higher temperatures, but the data contradicted this. A closer review of the literature revealed that temperature was not the sole factor influencing mercury levels. While higher temperatures accelerate bacteria conversion of mercury to methylmercury and increase fish metabolism, the higher average mercury levels in spring can be attributed to fish spawning cycles, which occur during this season. Spawning increases metabolic rates, leading to elevated mercury levels (Braaten et al., 2014). Furthermore, fish often experience accelerated growth during the summer, due to increased food ability,  which can dilute mercury concentrations. This explains the highest levels being in spring as opposed to summer.

Furthermore, we discovered that the land use region significantly impacted average mercury concentration.  Particularly the order of average mercury concentration by location from highest to lowest was Red Deer, Lower Athabasca, South Saskatchewan, North Saskatchewan, Upper Athabasca and Lower Peace. It was noted that Red Deer, Lower Athabasca, South Saskatchewan and North Saskatchewan exhibited the widest boxes, with similar ranges and distributions. This was an expected result due to the geographical connectedness of these regions as neighboring regions are likely affected by common environmental causes.

In terms of health related findings, we discovered that  20.8% of fish samples exceeded the threshold, while 70.2% were safe for consumption. This result was unexpected, as initially our team hypothesized that there would be very few samples that exceeded the threshold.

Finally, we also observed that mercury levels varied by waterbody type. Reservoirs had the highest concentrations, followed by lakes, rivers, stormwater ponds, and canals. This pattern was unexpected as we did not make specific predictions on how waterbody types would impact the concentration. However these results can easily be explained. Methylmercury formation occurs in anaerobic environments, such as those found in the sediments of stratified waters (Eckley et al., 2021). Reservoirs and lakes often trap organic matter and maintain these anoxic conditions, while rivers and canals experience more mixing and oxygenation, limiting methylmercury production (Baldwin et al., 2022). Stormwater ponds typically have lower mercury concentrations due to high turnover rates and frequent mixing, which disrupt conditions favorable for methylation (Baldwin et al., 2022).
